## Flow of git



# Basics Mostly used commands in Git

#### For making repository
``` 
git init
```

#### For Cloning repository
- copy code url to copy others repository
```
git clone https://github.com/aniketjayant953/sample-git-app.git 
```

#### For moving the fill in the staging index
- if you feel like you did some important work and the file needs to be tracked so you add the file
```
git add app.py
```

#### For moving all files to staging index
```
git add.
```

#### For checking Status
- shows branch
- shows you are upto date with branch master
- changes to be commited for files that you just added
```
git status
```

#### For excluding a particular file and track all files
- make a '.gitignore' file and the name of the file in that directory
- if you run 'git status' it will not show that particular file now but will show gitignore file



#### For commiting chages
- commits the file that are in staging index to be uploaded
- label should be written as suffix to 'This commit will'
```
git commit -m 'add sidebar to the website'
```

#### To view all commits
- it will shows summary for all commits
- it contains a 'SHA(simple hashing algo)' a unique name for that commit file
```
git log
```

#### To view commits in one line
```
git log --oneline
```

#### To see kitne line me kitne changes hue
- **'filename' | 1+**     -> this means 1 line added
- **'filename' | 3++-**   -> this means 2 lines added 1 removed
    
```
git log --stat
```

#### To escape vim editor when you ran above command
- scroll down till you see -> '(END)' and press 'Q' to escape

#### To see more detailed commits 
```
git log -p
```

#### To see particular files changes 
- green color will show added lines
- red color will show removed lines
- '@@ -0,0 +1,12 @@' -> this line will tell how many lines were there before commit and after commit
```
git show (SHA of that file)
```

#### To see the changes that you haven't yet added or commited right from the source file
```
git diff
```

# Version Control
### Version Tag -- X.Y.Z
**X** - The major version used for making major and backword compatible changes<br>
**Y** - The minor version, used for adding functionality while maintaing backwards compatibility<br>
**Z** - The patch version, used for making small bug fixes while mainitaing backwards compatibility


#### For giving version tag to a commit or a file
```
git tag -a v1.0.0
```
- you will greeted with a vim editor
- press Fn + I and it will allow to type 'v1.0.1'
- press 'escape' bring cursor in the bottom and click and then type ':wq' and press enter
- to check if the tag is placed correctly type 'git log --oneline'

#### To delete the tag
```
git tag -d v1.0.1
```
#### To give a tag to a particular file or a past commit
- enter that particular file
```
git show (SHA of a file)
```
- and then type the above code



## Branches

#### To create Branches

```
git branch sidebar
```
- **head-> master** -> this indicates that the branch is currently on master and head is the latest committed file


-  Run this and the asterick sign will tell you which branch you are currently on 
```
git branch
```


#### To jump to a branch
```
git checkout sidebar
```

#### To go on to a specific versions of a past file to make branches
```
git branch 'branch name' 'SHA of that file'
```
- run git log --oneline to get SHA of that file

#### To see all the commits in brief

```
git log --oneline --all
```

#### To see a graphical representation of how the branches are created

```
git log --oneline --all --graph
```

#### To delete a branch

```
git branch -d sidebar
```
- if the branch is not merged with the master branch and you still want to delete it just add capitial '-D' in place of '-d'
- you can't only delete the branch which you are not on



# Merging

### Two type of merging
1. Fast Forward Merging
    - If the branch is created on master and no further changes has been made to master branch but a lot of changes has been made to the cut out branch the merging will be fast forwarded to the new branch and the head will be latest commit made to the new branch
  
  
2. Regular Merging
    - This is normal merging




#### In case of Fast Forward to merge with the other branch


``` 
git merge sidebar
```
- You have to be on master branch to run this code


#### In case of Regular Merging to merge with the other branch

``` 
git merge sidebar
```
- This will open the vim editor just press ':wq' and 'Enter'

# Conflict

- Sometimes while merging different branches have different code at the same line in order to solve this git raises an error showing conflict between 2 file now its upto you which one to select

- git adds indicator in the code file showing the point of conflict

```
<<<<<<<< Head 
                    --> Here will be the code of master branch
                    
============
                    --> Here will be the code of sidebar branch

>>>>>>>>sidebar
```

- Now you have to choose which code to keep and which to trash
- After decision remove the code which is to be thrown away from the code file and also remove the special character '<< head == >>sidebar' and just do normal commit to git bash

## To PUSH the local repository to remote repository

#### First make the remote repository on github

#### After the repository is created
```
git remote add origin `repository link`
```
- You can find the repository link after the repo is created
#### To push the local file to remote 
```
git push origin master
```

#### To see if the local and remote repo is in sync
- If you run 'git log --online' and the 'HEAD' is on 'master' as well as 'master/origin' that means the local and remote is on the same page 


#### To PULL a file from remote repo
```
git pull origin master
```
